```python
import pandas as pd
import numpy as np

# Load the flight_data.pkl dictionary
# This path was confirmed previously.
df = pd.read_pickle("/content/drive/MyDrive/ngafid_data/2days/2days/flight_data.pkl")
print(f"Loaded flight_data.pkl: Type {type(df)}, Number of flights: {len(df)}")

# Re-build flight-level features from each NumPy array using summary statistics
# This logic was previously in cell eu1dF92NKsRG.
rows = []
for flight_id, arr in df.items():
    arr = np.array(arr, dtype=float)
    # Ensure arr is not empty before calculating statistics
    if arr.size > 0:
        means = np.nanmean(arr, axis=0)
        stds = np.nanstd(arr, axis=0)
        mins = np.nanmin(arr, axis=0)
        maxs = np.nanmax(arr, axis=0)
    else:
        # Handle empty arrays by assigning NaN or appropriate default values
        # Assuming 23 sensors based on previous output (4723, 23)
        num_sensors = 23 # Inferring from previous output `shape (4723, 23)`
        means = np.full(num_sensors, np.nan)
        stds = np.full(num_sensors, np.nan)
        mins = np.full(num_sensors, np.nan)
        maxs = np.full(num_sensors, np.nan)

    row = {"flight_id": flight_id}
    for i in range(len(means)):
        row[f"sensor_{i}_mean"] = means[i]
        row[f"sensor_{i}_std"] = stds[i]
        row[f"sensor_{i}_min"] = mins[i]
        row[f"sensor_{i}_max"] = maxs[i]
    rows.append(row)

features_df = pd.DataFrame(rows)

print(f"\nShape of features_df: {features_df.shape}")
print("Columns of features_df:")
print(features_df.columns.tolist()[:5], '...', features_df.columns.tolist()[-5:])
print("\nFirst 5 rows of features_df:")
display(features_df.head())

# Now, re-attempt the merge
# Check the data types of the key columns before merging
print("\nData type of 'Master Index' in header_df:", header_df['Master Index'].dtype)
print("Data type of 'flight_id' in features_df:", features_df['flight_id'].dtype)

# Perform an inner merge between header_df and features_df on the common flight ID column
modeling_df = pd.merge(header_df, features_df, left_on='Master Index', right_on='flight_id', how='inner')

print(f"\nShape of the merged DataFrame (modeling_df): {modeling_df.shape}")
print("Columns of the merged DataFrame (modeling_df):")
print(modeling_df.columns.tolist()[:5], '...', modeling_df.columns.tolist()[-5:])

print("\nFirst 5 rows of the merged DataFrame (modeling_df):")
display(modeling_df.head())
```

```python
# Create the binary target variable 'maintenance_within_2days'
# 1 if 'before_after' is 'before' and 'date_diff' is -2, -1, or 0
# 0 otherwise
modeling_df['maintenance_within_2days'] = (
    (modeling_df['before_after'] == 1) & (modeling_df['date_diff'] >= -2)
).astype(int)

print("First 5 rows of modeling_df with new target column:")
display(modeling_df[['Master Index', 'before_after', 'date_diff', 'maintenance_within_2days']].head())
```

```python
# Show the class balance of the new target variable
target_counts = modeling_df['maintenance_within_2days'].value_counts()
target_percentages = modeling_df['maintenance_within_2days'].value_counts(normalize=True) * 100

print("Class balance for 'maintenance_within_2days':")
print(target_counts)
print("\nPercentage of each class:")
print(target_percentages)

# Display the full DataFrame info to confirm the new column and its type
print("\nUpdated modeling_df info:")
modeling_df.info()
```

```python
# Define features (X) and target (y)
# Drop 'flight_id' as it's redundant with 'Master Index' after merge, and 'Master Index' itself if not needed as a feature.
# Drop original label columns and 'before_after', 'date_diff' which were used to create the target.
# Also drop 'hierarchy' as it has many nulls and 'label' (original string label).

# Identify columns to drop from features. These are meta-data or directly related to the target.
columns_to_drop = [
    'Master Index', # Often an ID column, not a predictive feature
    'flight_id',    # Redundant ID column from features_df
    'before_after', # Directly used to create target
    'date_diff',    # Directly used to create target
    'label',        # Original string label, not needed for training
    'hierarchy',    # Too many null values, not suitable as a direct feature without imputation/encoding
    'class',        # Another form of label, often numerical encoding of 'label'
    'target_class', # Another form of label
    'hclass',       # Another form of label
    'fold',         # Likely for cross-validation splitting, not a feature itself
    'maintenance_within_2days' # This is our target variable
]

# Ensure only existing columns are dropped to avoid errors
existing_columns_to_drop = [col for col in columns_to_drop if col in modeling_df.columns]

X = modeling_df.drop(columns=existing_columns_to_drop)
y = modeling_df['maintenance_within_2days']

print(f"Shape of features (X): {X.shape}")
print(f"Shape of target (y): {y.shape}")
print("\nFirst 5 rows of features (X):")
display(X.head())
print("\nFirst 5 values of target (y):")
display(y.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Split the data into training and testing sets
# Using a 70/30 split, stratifying by 'y' to maintain class balance in both sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")

# Initialize and train the RandomForestClassifier
# Using class_weight='balanced' is a good practice for binary classification, even with relatively balanced classes,
# to give slightly more importance to the minority class if it exists or when performance is critical.
# Set random_state for reproducibility.
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')

print("\nTraining RandomForestClassifier...")
model.fit(X_train, y_train)
print("RandomForestClassifier training complete.")
```

```python
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Make predictions on the test set
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1] # Probability of the positive class

# Calculate evaluation metrics
roc_auc = roc_auc_score(y_test, y_pred_proba)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"\nModel Evaluation on Test Set:")
print(f"ROC-AUC: {roc_auc:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
display(pd.DataFrame(cm, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1']))

# Plot Confusion Matrix
fig, ax = plt.subplots(figsize=(6, 6))
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Maintenance', 'Maintenance']).plot(ax=ax)
ax.set_title('Confusion Matrix')
plt.grid(False)
plt.show()
```

Based on the class balance:
- If the distribution is heavily skewed (e.g., one class is less than 10-15% of the data), it is considered imbalanced.
- For imbalanced datasets, common strategies include:
    - **`class_weight='balanced'`**: Many scikit-learn models (like `RandomForestClassifier` or `LogisticRegression`) have a `class_weight` parameter. Setting it to `'balanced'` automatically adjusts weights inversely proportional to class frequencies.
    - **Resampling techniques**: This involves either oversampling the minority class (e.g., SMOTE) or undersampling the majority class. However, these techniques should be applied carefully, typically after splitting the data into training and test sets to avoid data leakage.

We will proceed with model training and evaluate if imbalance handling is necessary based on the model's performance.
